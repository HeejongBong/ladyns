# 2. beta oscillation filter

```matlab
k = 2; % repeat for k=1,2,3
```

```matlab
fprintf('start cwt on v4 at k = %d\n',k);
v4_lfpb = zeros(1000, size(v4_lfp,2), size(v4_lfp,3));
for l = 1:1000
    fprintf('v4_%d_%d\n',k,l);
    i = (k-1) * 1000 + l;
    for j = 1:size(v4_lfp, 2)    
        [wt, f] = cwt(squeeze(v4_lfp(i,j,:)), 'amor', 1e3, ...
                      'FrequencyLimits', [14.0, 18.0], 'VoicesPerOctave', 4);
        v4_lfpb(l,j,:) = wt(1,:);
    end
end
```

```matlab
fprintf('start cwt on pfc\n');
pfc_lfpb = zeros(1000, size(pfc_lfp,2), size(pfc_lfp,3));
for l = 1:1000
    fprintf('pfc_%d_%d\n',k,l);
    i = (k-1) * 1000 + l;
    for j = 1:size(pfc_lfp, 2)    
        [wt, f] = cwt(squeeze(pfc_lfp(i,j,:)), 'amor', 1e3, ...
                      'FrequencyLimits', [14.0, 18.0], 'VoicesPerOctave', 4);
        pfc_lfpb(l,j,:) = wt(1,:);
    end
end
```

```matlab
save(strcat(save_dir, 'lfp_beta_',k,'.mat'), 'v4_lfpb', 'pfc_lfpb')
```

## EDA on correlation

```matlab
v4_lfpb_corr = corrcoef(reshape(permute(abs(v4_lfpb), [1 3 2]), [], size(v4_lfpb,2)));

v4_lfpb_corr_i = zeros([10 10]);
for i = 1:10
    for j = 1:10
        if ~isnan(maps.v4(i,j))
            v4_lfpb_corr_i(i,j) = v4_lfpb_corr(1, maps.v4(i,j));
        end
    end
end
```

```matlab
figure

subplot(1,2,1)
imagesc(v4_lfpb_corr)
caxis([-1 1])
axis square
colormap rdbu, colorbar

subplot(1,2,2)
imagesc(v4_lfpb_corr_i)
caxis([-1 1])
axis square
colormap rdbu, colorbar
```

```matlab
pfc_lfpb_corr = corrcoef(reshape(permute(abs(pfc_lfpb), [1 3 2]), [], size(pfc_lfpb,2)));

pfc_lfpb_corr_i = zeros([10 10]);
for i = 1:10
    for j = 1:10
        if ~isnan(maps.pfc(i,j))
            pfc_lfpb_corr_i(i,j) = pfc_lfpb_corr(1, maps.pfc(i,j));
        end
    end
end
```

```matlab
figure

subplot(1,2,1)
imagesc(pfc_lfpb_corr)
caxis([-1 1])
axis square
colormap rdbu, colorbar

subplot(1,2,2)
imagesc(pfc_lfpb_corr_i)
caxis([-1 1])
axis square
colormap rdbu, colorbar
```
